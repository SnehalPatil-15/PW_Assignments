**Question 1 : What is the difference between a function and a method in Python?**

**ANS 1]** FUNCTIONS :
A function is a solitary block of code that completes a specified job. You declare it using the def keyword and may access it from anywhere in your code.
Functions accept input parameters, conduct operations, and can return outputs. They are reusable, so you may use them several times throughout your program.

METHOD : A method is similar to a function, except it is connected with an object (often a class instance). Methods are defined within a class and frequently operate on the data stored within that class.
Methods are accessed using dot notation, which implies you call them on an object.

```python
#FUNCTION

def add(a, b):
    return a + b

result = add(3, 5)
print(f"The Output of function : {result}")
print()

#METHOD
class Calculations:
    def add(self, a, b):
        return a + b

    def sub(self, a, b):
        return a - b

cal = Calculations()
result1 = cal.add(3, 3)
print(f"The Output of method : {result1}")
result2 = cal.sub(5, 3)
print(f"The Output of method : {result2}")
```

**Question 6 : Explain the concept of generators in Python and how they are defined.**

**ANS 6]** When for any iterable we wish to generate result one by one rather than calculating all values and then returning them all together we use generators.

In generator function is like any other function but with one difference, instead of using 'return' keyword we make use of 'yield' keyword.

When this function encounters yield statement, it returns the value specified in yield and pauses its execution state.
The next time this generator is called it resumes from where it left off.


```python
#Example of generator function
def square_generator(n):
    for i in range(n):
        yield i ** 2

Sq_gen = square_generator(5) #Creating generator object
print(Sq_gen) #Printing generator object
print(next(Sq_gen)) #Printing values one by one
print(next(Sq_gen))
print(next(Sq_gen))
print(next(Sq_gen))
print(next(Sq_gen))
```

**Question 7 : What are the advantages of using generators over regular functions?**

**ANS 7]** Regular function performs calculations on an iterable and returns it in one go. In this case, memory usage is more since function would require that particular iterable to store the result. Execution time is also more because the function spends time on calculating all the values and storing it in iterable, finally entire result is returned altogether.

Generator function returns value one by one which means it would calculate result for the first value and immediately return that value. This reduces the use of memory space and the execution time of the table.

Hence, generator funtion in python is a way to create iterators in more concise and memory efficient manner

```python
#Example of regular function vs generator function

print("Output of regular function")
def add_value():
  list1 = [1,2,3]
  result_list = []
  for i in list1:
    result_list.append(i+10) #will append all the values to list and then return list
  return result_list

print(add_value())

print("Output of generator function")
def add_value_gen():
  list1 = [1,2,3]
  for i in list1:
    yield i+10   #will return values one by one instead of returning the entire list

gen = add_value_gen()
print(next(gen))
print(next(gen))
print(next(gen))
```

**Question 10 : What is difference between 'map()', 'reduce()' and 'filter()' functions in Python?**

**ANS 10]** Map Function:

It maps function with an iterable passed to it. i.e On each value inside the iterable map() function applies the functionality that is passed to it.

Reduce Function:

Reduce function implements mathematical technique known as reduction on an iterable. It is a process where list of items is reduced to single cumulative value.

Filter function:

Filter function is used to filter out the elements from an iterable based on some condiion.

```python
#Example of map, reduce and filter functions
import functools

list1 = [1,2,3]
print("Original List 1: ", list1)
print("List 1 after applying map(): ",list(map(lambda x: x**2,list1)))
print("List 1 after applying reduce(): ",functools.reduce(lambda x, y: x+y,list1))
print("List 1 after applying filter(): ",list(filter(lambda x: x%2!=0,list1)))
```

**Question 1: Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**

**ANS 1]**

```python
def add_even(my_list):
  sum = 0
  for i in my_list:
    if i%2==0:
      sum = sum +i
  return sum

my_list = [3,5,8,2,8,7,9]
print("The list: ",my_list)
sum = add_even(my_list)
print("The addition of even numbers in list is: ",sum)
```

**Question 5: Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.**

**ANS 5]**

```python
class FibonacciIterator:
  def __init__(self, n):
    self.n = n #Number of fibinacci numbers to generate
    self.a, self.b = 0,1
    self.count = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.count < self.n:
      current = self.a
      self.a, self.b = self.b, self.a + self.b
      self.count = self.count+1
      return current
    else:
      raise StopIteration

n = int(input("For how many numbers do you want to generate fibonacci series?"))
fib = FibonacciIterator(n)

for i in fib:
  print(i)
```

```python
class FibonacciIterator:
    def __init__(self, n):
        """Initialize the iterator with the number of terms (n)."""
        self.n = n  # Number of Fibonacci numbers to generate
        self.a, self.b = 0, 1  # Starting values for the Fibonacci sequence
        self.count = 0  # Keeps track of how many numbers we have generated

    def __iter__(self):
        """Return the iterator object itself."""
        return self

    def __next__(self):
        """Generate the next Fibonacci number."""
        if self.count < self.n:
            # The next number is stored in 'a' and then we update 'a' and 'b'
            current = self.a
            self.a, self.b = self.b, self.a + self.b  # Update to the next Fibonacci numbers
            self.count += 1  # Increment the counter
            return current
        else:
            # If we have generated 'n' Fibonacci numbers, stop the iteration
            raise StopIteration

# Create an instance of FibonacciIterator to generate the first 10 Fibonacci numbers
n_terms = int(input("How many Fibonacci numbers would you like? "))
fib_iter = FibonacciIterator(n_terms)

# Print the Fibonacci sequence using the iterator
for number in fib_iter:
    print(number)
```

**Question 6: Write a generator function in Python that yields the powers of 2 up to a given exponent**

**ANS 6]**

```python
def power_generator(num):
  for i in range(num+1):
    result = 2 ** i
    yield result

num = int(input("Up to which exponent do you widh to generate powers of 2?"))
ans=power_generator(num)

for i in ans:
  print(i)
```

**Question 7: Implement a generator function that reads a file line by line and yields each line as a string**

**ANS 7]**

```python
#Creating a file
with open('StudentInfo.txt','w') as file:
  file.write("Name: Spriya Sikder; Class: 12th; Group: PCB\n")
  file.write("Name: Simran Kagale; Class: 12th; Group: PCM\n")
  file.write("Name: Ana; Class: 12th; Group: PCMB\n")

#Generator function that reads lines
def read_lines(StudentInfo):
    with open(StudentInfo,'r') as file:
        for line in file:
            yield line.strip()

for line in read_lines('StudentInfo.txt'):
    print(line)
```
